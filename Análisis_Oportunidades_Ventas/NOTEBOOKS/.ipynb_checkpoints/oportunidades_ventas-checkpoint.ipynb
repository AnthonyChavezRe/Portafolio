# 2.- Tranformación: Limpieza y Análisis del dataset.

```python
# Cambiamos el tipo de dato a la columna 'FechaCierre' ya que originalmente es de tipo 'object'
data['FechaCierre'] = pd.to_datetime(data['FechaCierre'])
```

```python
# identificar valores duplicados de la columna IdOportunidad, en esta columna no debe de haber datos duplicados ya que 
# el Id debe de ser unico para garantizar integridad de los datos.
data[data.duplicated(subset='IdOportunidad', keep=False)]
```

```python
# Se observa que hay valores de texto.
# reemplazamos 'cero' por '0', 'diez' por '10', 'cien' por '100', 'diez mil' por '10000' de la columna Importe.
def replace_values(data, column):
    data[column] = data[column].str.lower().str.strip()
    replace_val = {'cero': 0, 'diez': 10, 'cien': 100, 'diez mil': 10000}
    data[column] = data[column].replace(replace_val)
    return data
```

```python
# reemplazamos 'sin datos' por '0' en la columna 'Participantes', ya que es columna númerica.
def replace_val_sin_datos(data, column):
    data[column] = data[column].replace({'Sin datos' : 0})
    return data
```

```python
# función para imputar datos faltantes de la tabla 'zona' se imputan tomando encuanta la Moda 
# de la tabla ya que es una columna categorica y analizando los datos es viable usar la Moda.
def fill_na_in_column(data, column, fill_value):
    data[column] = data[column].fillna(fill_value)
    return data
```

```python
# funcion para eliminar duplicados de la columna 'IdOportunidad' ya que en esta columna son llaves primarias y no debe de haber valores duplicads.
def remove_duplicates(data, column):
    data = data.drop_duplicates(subset = column, keep = 'first')
    return data
```

```python
# ejecucion pipeline de transformación de datos
def preprocess_data(data):
    return(
        data.pipe(replace_values, 'Importe')
            .pipe(replace_val_sin_datos, 'Participantes')
            .pipe(fill_na_in_column, 'Zona', data['Zona'].mode().iloc[0])
            .pipe(remove_duplicates, 'IdOportunidad')
    )
```

```python
data = preprocess_data(data)
```

```python
# Normalizar a mayusculas 
data['Zona'] = data['Zona'].str.upper()
```

```python
# Quitar espacios
columns = ['IdOportunidad','IdEmpresa','IdPropietario','TipoDivisaAjuste','Zona']
for col in columns:
    data[col] = data[col].str.strip()
```

```python
# Cambiamos el tipo de dato a la colimna Importe a int
data['Importe'] = data['Importe'].astype('int64')
data['Participantes'] = data['Participantes'].astype('int64')
```

```python
# Solo vamos a manejar cantidades en moneda Nacional MXN
# Agregamos una Columna de Tipos de Cambio que relacione las divisas con su tipo de cambio a MXN:
tipo_de_cambio = {
    'MXN' : 1.00,
    'USD' : 20.20,
    'EUR' : 21.20
}
# Crear una columna con el tipo de cambio según la divisa
data['TipoCambio'] = data['TipoDivisaAjuste'].map(tipo_de_cambio)

# Calcular el importe en MXN
data['ImporteEnMXN'] = data['Importe'] * data['TipoCambio']
```

```python
# Mostramos los primeros q0 registros del DataFrame
data.head(10)
```

```python
# Se calsifican las Zonas en dos grupos 'Central' y 'Periferica'
def clasificar_zona(Zona):
    if Zona in ['ZONA 1','ZONA 2']:
        return 'Central'
    elif Zona in ['ZONA 3','ZONA 4','ZONA 5']:
        return 'Periferica'
    return 'Desconocida'

# Se  crea la columna 'ClasificacionZona'
data['ClasificacionZona'] = data['Zona'].apply(clasificar_zona)
```

```python
# Crear rangos de Importe
def clasificat_importe(importante):
    if 0 <= importante < 50000:
        return 'Muy Bajo'
    elif 5000 <= importante < 1000000:
        return 'Bajo'
    elif 1000000 <= importante < 10000000:
        return 'Medio'
    elif 10000000 <= importante < 20000000:
        return 'Alto'
    else:
        return 'Muy Alto'
data['RangoImporte'] = data['ImporteEnMXN'].apply(clasificat_importe)
```

```python
# Calcular rendimiento promedio por zona sumando el Importe y dividiéndolo entre el número de Participantes. 
# Agregaremos una columna derivada RendimientoPorZona:
data['RendimientoPorZona'] = data['ImporteEnMXN'] / data['Participantes'].replace(0, 1)
```

```python
# Podemos derivar columnas como el año o el mes de la columna
# Extraer año y mes de la fecha de cierre
data['AñoCierre'] = pd.to_datetime(data['FechaCierre']).dt.year
data['MesCierre'] = pd.to_datetime(data['FechaCierre']).dt.month
```

```python
# Calcular el importe acumulado para cada propietario
data['ImporteAcumulado'] = data.groupby('IdPropietario')['ImporteEnMXN'].cumsum()

# filtrar propietarios con importe inicial mayor a 0
data = data[data.groupby('IdPropietario')['ImporteEnMXN'].transform('first') > 0]

# Calcular el crecimiento acumulado relativo al importe inicial
data['CrecimientoRelativo'] = data['ImporteAcumulado'] / data.groupby('IdPropietario')['ImporteEnMXN'].transform('first') - 1
```

```python
# Agrupaciones por Zona: Permite obtener insights como total de importes por zona.
agrupacion_zona = data.groupby(['Zona', 'ClasificacionZona'])['ImporteEnMXN'].sum().reset_index()
agrupacion_zona.rename(columns={'ImporteEnMXN': 'TotalImportePorZonaMXN'}, inplace=True)

# Ordenamos de mayor a menor.
agrupacion_zona = agrupacion_zona.sort_values(by = 'TotalImportePorZonaMXN', ascending = False)
print(agrupacion_zona)
```

# 3.- Carga de dataset Transformado y normalizado.

Cargar DataFrame en una base de datos en SQL Server Management Studio (SSMS) desde Python utilizando la biblioteca pyodbc o sqlalchemy.

```python
import pyodbc

# Conexión a tu SQL Server con seguridad integrada
conn = pyodbc.connect(
    "DRIVER={ODBC Driver 17 for SQL Server};"
    "SERVER=DESKTOP-35197TM\\SQLEXPRESS;"
    "DATABASE=Oportunidadesdb;"
    "Trusted_Connection=yes;"
)

cursor = conn.cursor()

# Consulta para crear la tabla
create_table_query = """
CREATE TABLE Oportunidades (
    IdOportunidad NVARCHAR(50) PRIMARY KEY,
    IdEmpresa NVARCHAR(50),
    IdPropietario NVARCHAR(50),
    TipoDivisaAjuste NVARCHAR(50),
    Zona NVARCHAR(50),
    Importe INT,
    FechaCierre DATE,
    Participantes INT,
    TipoCambio FLOAT,
    ImporteEnMXN FLOAT,	
    ClasificacionZona NVARCHAR(50),
    RangoImporte NVARCHAR(50),
    RendimientoPorZona FLOAT,
    AñoCierre INT,
    MesCierre INT,
    ImporteAcumulado FLOAT,
    CrecimientoRelativo FLOAT
);
"""

try:
    cursor.execute(create_table_query)
    conn.commit()
    print("✅ Tabla 'Oportunidades' creada correctamente")
except pyodbc.Error as e:
    print("❌ Error al crear la tabla:", e)
finally:
    cursor.close()
    conn.close()
```

# Carga del Dataframe

```python
from sqlalchemy import create_engine
import urllib

# Crear la cadena de conexión con urlencode para espacios y barras invertidas
params = urllib.parse.quote_plus(
    "DRIVER={ODBC Driver 17 for SQL Server};"
    "SERVER=DESKTOP-35197TM\\SQLEXPRESS;"
    "DATABASE=Oportunidadesdb;"
    "Trusted_Connection=yes;"
)

# Crear el motor de SQLAlchemy
engine = create_engine(f"mssql+pyodbc:///?odbc_connect={params}")

# Cargar el DataFrame al SQL Server
try:
    data.to_sql('Oportunidades', con=engine, if_exists='append', index=False)
    print("✅ Datos cargados exitosamente a la tabla 'Oportunidades'")
except Exception as e:
    print("❌ Error al cargar los datos:", e)
```
